En spelare $A$'s elo uppdateras enligt
$$ R'_A=R_A+K\cdot(S_A-E_A) $$
där $R'_A$ är den nya elon, $R_A$ är den gamla elon, $K$ är en multiplikationsfaktor, $S_A$ är matchens utfall (vinst: $S_A=1$, förlust: $S_A=0$) och $E_A$ är det förväntade utfallet baserat på spelarens elo och motspelarnas elo. 

$E_A$ beräknas enligt
$$ E_A = \frac{Q_A}{\sum_{i \in players}Q_i}$$
där $Q_i = 10^{R_i/400}$ och $R_i$ är den elo spelare $i$ har.
